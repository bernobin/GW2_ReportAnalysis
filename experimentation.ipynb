```python
import json
import pathlib
```

```python
fileName = '20230223-220133_sam_fail.json'
filePath = pathlib.Path('./reports')

file = pathlib.Path(filePath/fileName)

text = open(file)
data = json.load(text)
```

```python
for element in data:
    print(element)
```

```python
timers = {}

for phase_dict in data['phases']:
    timers[phase_dict['name']] = (phase_dict['end'] - phase_dict['start'])/1000
    
for t in timers:
    print(f"{t: <18} {timers[t]: >18}s")
```

```python
progress = data['targets'][0]['healthPercentBurned']

print(progress)
```
